# Data collected 4-22-2020

## Settings

```python
import csv
import pandas as pd
```

## Load file

```python
df = pd.read_csv('4-22-2020_DLsite_Games.csv', encoding='utf-8')
```

## Add an earnings column (sales * price)

```python
df['earnings'] = df['price'] * df['sales']
```

## Create data frames based on year

```python
# check for non-year values
df['release year'].unique()
```

## Prune data with wrong years

```python
# noticed that all with non-year values had NaN release months
# get all values where the release month isn't null 
# removes all incorrect years
df = df.loc[~df['release month'].isnull()] # delete

# check years again
all_year_vals = df['release year'].unique().tolist()
all_year_vals.sort()
df['release year'].unique()
# There are listings from 1997?!
```

```python
# create a dictionary with years as keys
df_by_year = {}
for i in all_year_vals:
    df_by_year[i] = df.loc[df['release year'] == i]

# so if I want to get data based on year, just do something like df_by_year['2020']
```
